Surface coverage of each species under various tau_max in algorithm 3 (lambda = 1)

```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
# plt.rc('text', usetex=True)
from arguments import get_args
args = get_args()
algo="alg3" 
taumax_list = ["ssa", "1e-08", "3e-08", "5e-08"]
datapath = "alg3_taumaxs_data/"
savepath = "24_2_13_alg3_taumaxs/"
if not os.path.exists(savepath):
    os.makedirs(savepath)
```

```python
for specie_idx in range(len(args.sim.species_names)):
    if args.sim._init_state[specie_idx] == 0: # ignore species that cannot stay on the surface
        continue

    for tau in taumax_list:
        path = os.path.join(datapath, f"alg3_taumax_{tau}")
        
        file_name = os.path.join(path, f"{algo}_size_{args.lattice_size}_sample_{args.num_samples}_{args.sim.species_names[specie_idx]}.txt")

        with open(file_name, 'r') as file:
            lines = file.readlines()

        for i, line in enumerate(lines): # read the list of coverages from file
            if "res in list" in line:
                list_line = lines[i+1]
                break
        try:
            result_list = eval(list_line)
        except Exception as e:
            print(f"error when evaluating list: {e}")

        for i, line in enumerate(lines): # read the list of timestamps from file
            if "time stamps in list" in line:
                list_line = lines[i+1]
                break
        try:
            timestamp_list = eval(list_line)
            # print(timestamp_list)
        except Exception as e:
            print(f"error when evaluating list: {e}")
        try:
            float(tau)
            lab="$\\tau_{max}=$"+tau
        except ValueError:
            lab=tau


        plt.plot(timestamp_list, result_list, label=lab)

    plt.legend()
    sp_name = args.sim.species_names[specie_idx]
    sp_name = sp_name[:-2] if sp_name.endswith("_s") else sp_name

    plt.xlabel("Time (s)")
    plt.ylabel("Coverage")
    plt.title(f"{sp_name} Surface Coverage Under Various $\\tau_{{max}}$")
    
    plt.savefig(os.path.join(savepath, f"{sp_name}.png"))
    plt.savefig(os.path.join(savepath, f"{sp_name}.pdf"))
    plt.show()
    plt.clf()
    
```
